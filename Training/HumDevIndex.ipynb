```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("../Dataset/h.csv.csv")
data.columns = data.columns.str.strip()

# First 20 rows
data1 = data.head(20)

# -----------------------------
# 1. Unique Countries
# -----------------------------
print("Unique Countries:")
print(data1["Country"].unique())

# -----------------------------
# 2. Choose correct HDI column (2021)
# -----------------------------
hdi_col = "Human Development Index (2021)"

# -----------------------------
# 3. Plot HDI trend (1990 vs 2021 style comparison)
# -----------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="HDI Rank (2021)",
    y=hdi_col,
    data=data1
)
plt.title("HDI Rank vs HDI (2021)")
plt.show()

# -----------------------------
# 4. Correlation Heatmap (ONLY numeric columns)
# -----------------------------
plt.figure(figsize=(12,6))

numeric_data = data1.select_dtypes(include=["number"])
sns.heatmap(numeric_data.corr(), annot=False, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
